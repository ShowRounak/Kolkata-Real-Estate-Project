```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cleaned_data_v3.csv')
df.drop(columns=['Unnamed: 0'],inplace=True)
df.head()
```

### property type vs Price

```python
sns.barplot(x=df['PROPERTY_TYPE'], y=df['price'], estimator=np.median)
```

```python
sns.boxplot(x=df['PROPERTY_TYPE'], y=df['price'])
```

### property Type vs AREA

```python
sns.barplot(x=df['PROPERTY_TYPE'], y=df['AREA'], estimator=np.median)
```

```python
sns.boxplot(x=df['PROPERTY_TYPE'], y=df['AREA'])
```

### property Type vs price_per_unit

```python
sns.barplot(x=df['PROPERTY_TYPE'], y=df['PRICE_PER_UNIT_AREA'], estimator=np.median)
```

```python
sns.boxplot(x=df['PROPERTY_TYPE'], y=df['PRICE_PER_UNIT_AREA'])
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(pd.crosstab(df['PROPERTY_TYPE'],df['CITY'].sort_index()),cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
avg_price_per_loc = df.groupby('Location')['price'].mean().reset_index()
avg_price_per_loc
```

```python
plt.figure(figsize=(5, 100))
sns.heatmap(avg_price_per_loc.set_index('Location')[['price']], annot=True, fmt=".2f", linewidths=.5, cmap=sns.cubehelix_palette(as_cmap=True))
plt.title('Average Price Area wise')
plt.xlabel('Average Price')
plt.ylabel('location')
plt.show()
```

```python
avg_price_per_sqft_loc = df.groupby('Location')['PRICE_PER_UNIT_AREA'].mean().reset_index()
avg_price_per_sqft_loc
```

```python
plt.figure(figsize=(5, 100))
sns.heatmap(avg_price_per_sqft_loc.set_index('Location')[['PRICE_PER_UNIT_AREA']], annot=True, fmt=".2f", linewidths=.5, cmap=sns.cubehelix_palette(as_cmap=True))
plt.title('Average Price/sqft Area wise')
plt.xlabel('Average Price/sqft')
plt.ylabel('location')
plt.show()
```
